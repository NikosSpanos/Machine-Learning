```python
from sklearn.metrics import mean_squared_log_error

def rmsle(real, predicted):
    sum=0.0
    for x in range(len(predicted)):
        if predicted[x]<0 or real[x]<0: #check for negative values
            continue
        p = np.log(predicted[x]+1)
        r = np.log(real[x]+1)
        sum = sum + (p - r)**2
    return (sum/len(predicted))**0.5

rmsle(y_test, new_pr)

#ert len(y_test) == len(new_pr)

#erms_to_sum = [(math.log(new_pr[i] + 1) - math.log(y_test[i] + 1)) ** 2.0 for i,pred in enumerate(new_pr)]
#sum(terms_to_sum) * (1.0/len(y_test)) ** 0.5
```
